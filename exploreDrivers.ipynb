```python
# Imports
import pandas as pd
```

```python
#Read Files
drivers = pd.read_csv("data/drivers.csv")
drivers_standings = pd.read_csv("data/driver_standings.csv")
```

```python
drivers.head()
pd.options.mode.chained_assignment = None 
```

```python
drivers_reduce=drivers[["driverId","driverRef","nationality"]]
drivers_reduce["name"]=drivers["forename"]+" "+ drivers["surname"]
drivers_reduce
```

```python
drivers_standings
#Firsts
#Seconds
#Thirds
#TotalPoints

#Best position
prueba=drivers_standings.loc[drivers_standings["driverId"]==4]
prueba.loc[prueba["position"]==1]
```

```python
#Total Points
drivers_standings.groupby("driverId")["points"].sum()
```

```python
#Count positions
drivers_positionsaux=drivers_standings.groupby(["driverId","position"])["position"].count().reset_index(name="count")
drivers_positionsaux
```

```python
#Merge dataframe to get drivers name
drivers_positions = drivers_reduce.merge(drivers_positionsaux, on='driverId', how='inner', suffixes=('_1', '_2'))
drivers_positions_reduce =drivers_positions[["name","position","count"]]
drivers_positions_reduce
```

```python
driver_podiums = drivers_positions_reduce[drivers_positions_reduce['position'].between(1, 3)]
driver_podiums
```
