```python
# Imports
import pandas as pd
#Read Files
constructors = pd.read_csv("data/constructors.csv")
results = pd.read_csv("data/constructor_results.csv")
races = pd.read_csv("data/races.csv")
```

```python
constructors_reduce = constructors[["constructorId","name"]]
constructors_reduce
```

```python
results_reduce = results[["constructorId", "points"]]
results_reduce
```

```python
#Constructor all points
constructors_points = constructors.merge(results_reduce, on='constructorId', how='inner', suffixes=('_1', '_2'))
constructors_points = constructors_points.groupby("name")["points"].sum().reset_index(name="points")
constructors_points 
```

```python
#Results with year
results_year = results.merge(races, on='raceId', how='inner', suffixes=('_1', '_2'))
results_year = results_year[["raceId","constructorId","year", "points"]]
results_year 
```

```python
constructors_points_year = constructors.merge(results_year, on='constructorId', how='inner', suffixes=('_1', '_2'))
constructors_points_year = constructors_points_year.groupby(["name","year"])["points"].sum().reset_index(name="points")
constructors_points_year
```

```python
#Write to csv
constructors_points.to_csv("data/outputs/constructors_points.csv", sep=',', index=False)
constructors_points_year.to_csv("data/outputs/constructors_points_year.csv", sep=',', index=False)
```
